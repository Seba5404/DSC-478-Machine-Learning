```python
#populate bank data into dataframe


bank_df = pd.read_csv("bank_data.csv", sep=',')
bank_df.head() #show data frame
```

```python
bank_df.shape
```

```python
Bank_data = bank_df.drop("pep", axis=1, inplace = False) #drop PEP 
Bank_data.drop("id", axis = 1, inplace=True) #drop ID not needed 


Pep_Bank_data = bank_df.pep
```

```python
Bank_data.head() #dataset without pep 
```

```python
Pep_Bank_data.head() #isolatd PEP data 
```

```python
Bank_data.replace(to_replace = ['YES', 'NO'], value = [1.0, 0.0], inplace=True) #replace yes and no with 1 and 0 
#1 being yes, 0 being no 
Bank_data.replace(to_replace = ['FEMALE', 'MALE'], value = [1.0, 0.0], inplace=True) #Female = 1 , Male = 0 

#get dummies for region 
Bank_data_Dummies = pd.get_dummies(Bank_data[["region"]])

#combine Dummies with orginal bank data 
Bank_data_Wd = Bank_data.join(Bank_data_Dummies)

Bank_data_Wd.drop("region", axis = 1, inplace=True) #drop region as not needed since we have dummies
Bank_data_Wd.head()
```

```python
#HW problem 2 A training/test data has been split 
from sklearn.cross_validation import train_test_split
Bank_Data_train, Bank_Data_test, Bank_Data_target_train, Bank_Data_target_test = train_test_split(Bank_data_Wd, Pep_Bank_data, test_size=0.2, random_state=33)
#split data into training and testing 80% training, 20% testing 

#HW problem 2 A training/test data has been split 
Bank_Data_train.head()
```

```python
Bank_Data_target_train.head() #target_train = pep column 
```

```python
#Problem 2 B) Run scikit-learn's KNN classifier on the test set. 
#Note: in the case of KNN, you should first normalize the data so that all 
#attributes are in the same scale (normalize so that the values are between 0 and 1).

from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler().fit(Bank_Data_train)
Bank_data_train_norm = min_max_scaler.transform(Bank_Data_train)
Bank_data_test_norm = min_max_scaler.transform(Bank_Data_test)
```

```python
Bank_data_train_norm[0:5] # showoutput that data has been normlaized 
```

```python
 Bank_data_test_norm[0:5] #output of test data normalized 
```

```python
#Next, we call the predict function on the test intances to produce the predicted classes.

knnpreds_test = knnclf.predict(Bank_data_test_norm)

#print out results 
print(knnpreds_test)

```

```python
from sklearn.metrics import classification_report
print(classification_report(Bank_Data_target_test, knnpreds_test)) 
#compare Accruary rating for actual test labels vs predicted 
```
